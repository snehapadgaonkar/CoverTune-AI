```python
# STEP 1: Install all required libraries
!pip uninstall -y whisper
!pip install git+https://github.com/openai/whisper.git
!pip install git+https://github.com/huggingface/transformers.git
!pip install diffusers torchaudio gradio accelerate
!pip install sentencepiece
!pip install langdetect

# STEP 2: Import Libraries
import whisper
import torch
from transformers import pipeline, MarianMTModel, MarianTokenizer
from langdetect import detect
from diffusers import StableDiffusionPipeline
from PIL import Image
import uuid

# STEP 3: Load Whisper model (for transcription)
audio_model = whisper.load_model("medium")

# STEP 4: Transcribe Audio File
def transcribe_audio(file_path):
    result = audio_model.transcribe(file_path)
    return result["text"]

# STEP 5: Detect Language
def detect_language(text):
    return detect(text)

# STEP 6: Translate if needed
def translate_text(text, source_lang):
    lang_code_map = {
        "ta": "ta-en",
        "hi": "hi-en",
        "ml": "ml-en",
        "te": "te-en",
        "kn": "kn-en"
    }
    if source_lang not in lang_code_map:
        return text  # already English

    model_name = f"Helsinki-NLP/opus-mt-{lang_code_map[source_lang]}"
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    tokens = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    translated = model.generate(**tokens)
    english = tokenizer.decode(translated[0], skip_special_tokens=True)
    return english

# STEP 7: Emotion Detection
emotion_classifier = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment")

def detect_emotion(text):
    emotions = emotion_classifier(text)
    return emotions[0]["label"]

# STEP 8: Generate Prompt for Album Cover
def generate_prompt(text, emotion):
    return f"A highly detailed album cover illustration, based on a song with the emotion: {emotion}. The lyrics are about: {text}. Artistic, surreal, colorful, with musical symbolism."

# STEP 9: Load Stable Diffusion
sd_pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
).to("cuda")

# STEP 10: Generate Album Cover Image
def generate_album_cover(prompt):
    image = sd_pipe(prompt).images[0]
    filename = f"album_cover_{uuid.uuid4().hex[:8]}.png"
    image.save(filename)
    return image

# STEP 11: Full Flow
def covertune_flow(audio_path):
    print("🔊 Transcribing...")
    lyrics = transcribe_audio(audio_path)

    print("🌐 Detecting Language...")
    lang = detect_language(lyrics)

    print("🌍 Translating if needed...")
    lyrics_en = translate_text(lyrics, lang)

    print("🧠 Analyzing Emotion...")
    emotion = detect_emotion(lyrics_en)

    print("📝 Generating Prompt...")
    prompt = generate_prompt(lyrics_en, emotion)

    print("🎨 Creating Album Cover...")
    image = generate_album_cover(prompt)

    print("✅ Done!")
    display(image)

# Example Usage:
# Upload an audio file first (MP3/WAV)
from google.colab import files
uploaded = files.upload()

for file_name in uploaded.keys():
    covertune_flow(file_name)
```
